```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
print("Features: ")
display(list(df.columns))

print('\nNumerical values: ')
display(list(df.select_dtypes(include='number').columns))

print('\nNominal values: ')
display(list(df.select_dtypes(include=['category','object']).columns))
```

```python
# Summary Statistics
stats = df.describe(include='all')
stats.loc['var'] = df.select_dtypes(include='number').var()
stats.loc['range'] = stats.loc['max'] - stats.loc['min']

print('\nStatstics of each feature: ')
display(stats.loc[['min', 'max', 'mean', 'range', 'std', 'var', '25%', '50%', '75%']])
```

```python
# Histogram for all features
import sklearn.preprocessing as preprocessing

le = preprocessing.LabelEncoder()
new_df = df.copy()
new_df['Species'] = le.fit_transform(new_df['Species'])

new_df.drop(labels=['Species'], axis=1).hist(figsize=(15, 15), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms for Each Numeric Feature", fontsize=16)
plt.tight_layout()
plt.show()

# Histogram for categorical feature: Species
plt.figure(figsize=(6, 4))
plt.hist(new_df['Species'], bins=3, color='skyblue', edgecolor='black')
plt.title('Distribution of Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2], labels=le.classes_)  # Replace 0,1,2 with original class names
plt.tight_layout()
plt.show()
```

```python
# Box plot for each feature
for column in list(df.columns):
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()

# Box plot of all features together (excluding id)
plt.figure(figsize=(12,8))
sns.boxplot(data=df.drop(labels=['Id'], axis=1))
plt.show()
```

```python
# Identifying outliers

for column in list(df.columns)[:-1]:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    if (outliers.size > 0):
        print(f"\nOutliers for {column}: ")
        display(outliers)
    else:
        print(f"\nNo outliers for {column}")
```
