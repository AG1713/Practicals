```python
# 1. Import all the required Python Libraries.
import pandas as pd
import numpy as np
```

```python
# 2. Locate an open source data from the web (e.g., https://www.kaggle.com). Provide a clear
#    description of the data and its source (i.e., URL of the web site).
# 3. Load the Dataset into pandas dataframe.
iris = pd.read_csv('Iris.csv')
```

```python
# Overview of data set
print("\nhead:")
display(iris.head())

print("\ntail:")
display(iris.tail())

print("\ncolumns:")
display(iris.columns)

print("\ninfo:")
display(iris.info)

print("\nshape:")
display(iris.shape)
```

```python
# Checking for missing values
print("Null values:")
display(iris.isnull())

print("\nMissing values per column:")
display(iris.isnull().sum())

print("\nTrue if any null in column:")
display(iris.isnull().any())

print("\nRows with any missing values:")
display(iris.isnull().any(axis=1))
```

```python
# Checking data types and variable categories
print("Data types of each column:")
display(iris.dtypes)

print("\nCategorical variables:")
display(iris.select_dtypes(include='object').columns)

print("\nNumerical variables:")
display(iris.select_dtypes(include='number').columns)
```

```python
# Summarize the data types of each column
print("Data types of each column:")
display(iris.dtypes)
```

```python
# Check for categorical variables (should be 'object' or 'category')
categorical_cols = iris.select_dtypes(include='object').columns
print("Categorical variables:")
display(categorical_cols)

# Check for numerical variables (should be 'int' or 'float')
numerical_cols = iris.select_dtypes(include='number').columns
print("\nNumerical variables:")
display(numerical_cols)
```

```python
# Last column should be category instead of object
df = iris
df['Species'] = df['Species'].astype('category')
df.dtypes

# For converting to numeric types,
# Example: Convert 'age' column to numeric if it's not already
# df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Check the result
display(df.head())
```

```python
# Categorical values to quantitative
import sklearn.preprocessing as preprocessing 
df1 = df2 = df

# Label encoding
print("\nLabel Encoding: ")
le = preprocessing.LabelEncoder()
df1['Species'] = le.fit_transform(df1['Species'])
display(df1)
display(df.dtypes)

# One hot encoding
print("\nOne hot Encoding: ")
df2 = pd.get_dummies(df2, columns=['Species'])
display(df2)
display(df2.dtypes)
```
