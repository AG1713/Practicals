```python
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', None) # Show full column contents (if applicable)
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
print("Grouped Summary")
display(df.groupby('Species').describe())
display(df.groupby('Species').agg(['mean', 'median', 'min', 'max', 'std']))
```

```python
print("Overall Statistics: ")
display(df.describe())

# For each species
print("\nStatistics for species Iris-setosa")
display(df[df['Species'] == 'Iris-setosa'].describe())

print("\nStatistics for species Iris-setosa")
display(df[df['Species'] == 'Iris-versicolor'].describe())

print("\nStatistics for species Iris-setosa")
display(df[df['Species'] == 'Iris-virginica'].describe())

df['Species'].unique()
```
