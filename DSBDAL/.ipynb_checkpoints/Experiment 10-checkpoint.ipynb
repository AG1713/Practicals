```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset from sklearn
iris_data = load_iris()

# Create DataFrame
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = iris_data.target
df['species'] = df['species'].apply(lambda x: iris_data.target_names[x])

# Display first few rows
print("ðŸ”¹ Sample of the Dataset:")
print(df.head())

# 1. List down the features and their types
print("\nðŸ”¹ Features and Their Data Types:")
print(df.dtypes)

print("\nðŸ”¹ Summary Statistics:")
print(df.describe())

# 2. Histogram for each feature
df.drop("species", axis=1).hist(figsize=(10, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("ðŸ”¹ Histograms for Each Numeric Feature", fontsize=16)
plt.tight_layout()
plt.show()

# 3. Boxplot for each feature
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x="species", y=column, data=df, palette='Set2')
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()

# Drop the 'species' column for quantile calculation and outlier detection
numeric_df = df.drop('species', axis=1)

# 4. Identify Outliers
print("\nðŸ”¹ Outlier Detection:")
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

# Count the number of outliers in each column
print(outliers.sum())

# Highlighting actual rows with outliers (optional)
print("\nðŸ”¹ Rows with Potential Outliers:")
print(df[outliers.any(axis=1)])

```

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# 1. List down the features and their types
print("\nFeatures and their Data Types:\n")
print(iris.dtypes)
print("\n")

# 2. Create a histogram for each feature
iris.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle('Histograms of Iris Features', fontsize=16)
plt.tight_layout()
plt.show()

# 3. Create a boxplot for each feature
for column in iris.columns[:-1]:  # Exclude 'species' (categorical)
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=iris[column], color='lightgreen')
    plt.title(f'Boxplot of {column}', fontsize=14)
    plt.xlabel(column)
    plt.show()

# 4. Compare distributions and identify outliers
print("\nOutlier Observations Based on Boxplots:")
for column in iris.columns[:-1]:
    Q1 = iris[column].quantile(0.25)
    Q3 = iris[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = iris[(iris[column] < lower_bound) | (iris[column] > upper_bound)]
    
    if not outliers.empty:
        print(f"- {column}: {len(outliers)} outliers found.")
    else:
        print(f"- {column}: No significant outliers.")
```
